```
# Full corrected script: Decision Tree on UCI Bank Marketing (bank-additional-full)##task3
# Copy/paste and run in one cell.
# Requirements: pandas, scikit-learn, matplotlib, requests, joblib
# If using Colab, you can run:
#   !pip install -U scikit-learn pandas matplotlib requests joblib

import os
import io
import zipfile
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import joblib

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix, roc_auc_score, roc_curve

# -----------------------
# 1) Download & load dataset
# -----------------------
zip_url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00222/bank-additional.zip"
print("Downloading dataset from UCI...")
r = requests.get(zip_url)
r.raise_for_status()

z = zipfile.ZipFile(io.BytesIO(r.content))
print("Files in zip:", z.namelist())

# choose the CSV (full or smaller)
csv_path = "bank-additional/bank-additional-full.csv"
if csv_path not in z.namelist():
    csv_path = "bank-additional/bank-additional.csv"

with z.open(csv_path) as f:
    df = pd.read_csv(f, sep=';')

print("Loaded dataframe shape:", df.shape)
print("Columns:", df.columns.tolist())

# -----------------------
# 2) Quick EDA & target
# -----------------------
# Target variable is 'y' (yes/no)  convert to 0/1
df['y'] = df['y'].map({'yes': 1, 'no': 0})
print("\nTarget distribution (counts):")
print(df['y'].value_counts())
print("\nTarget distribution (proportions):")
print(df['y'].value_counts(normalize=True))

# -----------------------
# 3) Prepare features / preprocessing
# -----------------------
# Remove 'duration' as recommended by dataset authors (leaks information).
if 'duration' in df.columns:
    df = df.drop(columns=['duration'])

X = df.drop(columns=['y'])
y = df['y']

# Identify numeric and categorical columns
numeric_cols = X.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_cols = X.select_dtypes(include=['object']).columns.tolist()

print("\nNumeric columns:", numeric_cols)
print("Categorical columns:", categorical_cols)

# ColumnTransformer: scale numeric, OneHotEncode categorical
preprocessor = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), numeric_cols),
        ("cat", OneHotEncoder(handle_unknown="ignore", sparse_output=False), categorical_cols),
    ],
    remainder='drop'
)

# -----------------------
# 4) Train/test split and Pipeline
# -----------------------
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42, stratify=y
)

clf = Pipeline(steps=[
    ("preproc", preprocessor),
    ("tree", DecisionTreeClassifier(random_state=42, max_depth=6, min_samples_leaf=20))
])

print("\nTraining Decision Tree (this may take a moment)...")
clf.fit(X_train, y_train)

# -----------------------
# 5) Evaluation
# -----------------------
y_pred = clf.predict(X_test)
y_pred_proba = clf.predict_proba(X_test)[:, 1] if hasattr(clf.named_steps['tree'], "predict_proba") else None

acc = accuracy_score(y_test, y_pred)
print(f"\nTest Accuracy: {acc:.4f}\n")
print("Classification Report:\n", classification_report(y_test, y_pred, digits=4))

# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion matrix:\n", cm)

# ROC AUC (optional  requires predict_proba)
if y_pred_proba is not None:
    auc = roc_auc_score(y_test, y_pred_proba)
    print(f"ROC AUC: {auc:.4f}")

# 5-fold CV accuracy
cv_scores = cross_val_score(clf, X, y, cv=5, scoring='accuracy', n_jobs=-1)
print("5-fold CV accuracy: mean={:.4f}, std={:.4f}".format(cv_scores.mean(), cv_scores.std()))

# -----------------------
# 6) Feature names after preprocessing & feature importances
# -----------------------
# Get transformed feature names
num_names = numeric_cols

ohe = clf.named_steps['preproc'].named_transformers_['cat']
try:
    # sklearn >= 1.0
    cat_names = ohe.get_feature_names_out(categorical_cols).tolist()
except Exception:
    # fallback: expand categories manually
    cat_names = []
    for col in categorical_cols:
        cats = X[col].astype(str).unique().tolist()
        cat_names += [f"{col}_{c}" for c in cats]

feature_names = num_names + cat_names

tree = clf.named_steps['tree']
importances = tree.feature_importances_
# pair and sort
fi = sorted(zip(feature_names, importances), key=lambda x: x[1], reverse=True)
print("\nTop 15 feature importances:")
for name, imp in fi[:15]:
    print(f"  {name}: {imp:.4f}")

# -----------------------
# 7) Plot a truncated tree for readability
# -----------------------
plt.figure(figsize=(20,10))
plot_tree(tree, feature_names=feature_names, class_names=['no','yes'],
          filled=True, rounded=True, max_depth=3, fontsize=8)
plt.title("Decision Tree (truncated to max_depth=3)")
plt.show()

# -----------------------
# 8) Optional: ROC curve plot (if probabilities available)
# -----------------------
if y_pred_proba is not None:
    fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
    plt.figure(figsize=(6,5))
    plt.plot(fpr, tpr, label=f'ROC AUC = {auc:.4f}')
    plt.plot([0,1],[0,1],'--', linewidth=0.7)
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve')
    plt.legend()
    plt.grid(True)
    plt.show()

# -----------------------
# 9) Save the trained pipeline
# -----------------------
os.makedirs("models", exist_ok=True)
model_path = "models/decision_tree_bank_marketing.pkl"
joblib.dump(clf, model_path)
print(f"\nSaved trained pipeline to {model_path}")

# -----------------------
# 10) Notes
# -----------------------
print("""
Notes:
- 'duration' was removed because it leaks information not known before the interaction ends.
- The tree hyperparameters (max_depth=6, min_samples_leaf=20) are simple regularization choices.
- For improved performance: try RandomForest / XGBoost, class_weight='balanced', or hyperparameter tuning (GridSearchCV).
- Consider resampling (SMOTE/undersampling) if you need to improve recall on the minority class.
""")
```

```
# -----------------------------------------------------------
# Task-05: Traffic Accident Data Analysis (FARS 2016 Format)
# Internship: Prodigy InfoTech
# Dataset: acc_16.csv.zip
# -----------------------------------------------------------

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import zipfile
import os

# ---------------- STEP 1: Extract & Load Dataset ----------------
zip_file = "acc_16.csv.zip"

if not os.path.exists("acc_16.csv"):
    with zipfile.ZipFile(zip_file, 'r') as z:
        z.extractall()
        print(" Zip file extracted successfully!")

# Load dataset
df = pd.read_csv("acc_16.csv", low_memory=False)
print(" Dataset Loaded Successfully!")
print(f"Shape: {df.shape}")
print("First 10 columns:", list(df.columns[:10]))

# ---------------- STEP 2: Select and Clean Key Columns ----------------
columns = ['URBANICITY', 'WEATHR_IM', 'LGTCON_IM', 'MAXSEV_IM', 'VE_TOTAL', 'REGION']
df = df[columns].dropna()

print("\n Selected columns for analysis:", columns)
print(df.head())

# ---------------- STEP 3: Map Encoded Columns ----------------
# URBANICITY (1=Rural, 2=Urban)
urban_map = {1: 'Rural', 2: 'Urban'}
df['URBANICITY'] = df['URBANICITY'].map(urban_map)

# Weather condition mapping (simplified example from NHTSA FARS)
weather_map = {
    1: 'Clear', 2: 'Rain', 3: 'Sleet/Hail', 4: 'Snow',
    5: 'Fog/Smog', 6: 'Severe Wind', 7: 'Blowing Sand/Dust',
    8: 'Other', 9: 'Unknown'
}
df['WEATHR_IM'] = df['WEATHR_IM'].map(weather_map).fillna('Unknown')

# Light condition mapping
light_map = {
    1: 'Daylight', 2: 'Dark - Lighted', 3: 'Dark - Unlighted',
    4: 'Dawn', 5: 'Dusk', 6: 'Other', 7: 'Unknown'
}
df['LGTCON_IM'] = df['LGTCON_IM'].map(light_map).fillna('Unknown')

# Region mapping (simplified based on FARS regions)
region_map = {
    1: 'Northeast', 2: 'Midwest', 3: 'South', 4: 'West'
}
df['REGION'] = df['REGION'].map(region_map).fillna('Unknown')

# ---------------- STEP 4: Visualization ----------------
sns.set(style="whitegrid")

# 1 Urban vs Rural Accidents
plt.figure(figsize=(6,4))
sns.countplot(x='URBANICITY', data=df, palette='Set2')
plt.title(' Accidents by Area Type')
plt.xlabel('Area')
plt.ylabel('Number of Accidents')
plt.tight_layout()
plt.show()

# 2 Weather Conditions
plt.figure(figsize=(8,5))
sns.countplot(y='WEATHR_IM', data=df, order=df['WEATHR_IM'].value_counts().index, palette='coolwarm')
plt.title(' Accidents by Weather Condition')
plt.xlabel('Number of Accidents')
plt.ylabel('Weather Condition')
plt.tight_layout()
plt.show()

# 3 Light Conditions
plt.figure(figsize=(8,5))
sns.countplot(y='LGTCON_IM', data=df, order=df['LGTCON_IM'].value_counts().index, palette='magma')
plt.title(' Accidents by Light Condition')
plt.xlabel('Number of Accidents')
plt.ylabel('Light Condition')
plt.tight_layout()
plt.show()

# 4 Accident Severity Distribution
plt.figure(figsize=(6,4))
sns.countplot(x='MAXSEV_IM', data=df, palette='plasma')
plt.title(' Distribution of Accident Severity')
plt.xlabel('Severity Level Code')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

# 5 Vehicles Involved
plt.figure(figsize=(6,4))
sns.histplot(df['VE_TOTAL'], bins=15, kde=True, color='skyblue')
plt.title(' Distribution of Number of Vehicles per Accident')
plt.xlabel('Number of Vehicles')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

# 6 Accidents by U.S. Region
plt.figure(figsize=(7,4))
sns.countplot(x='REGION', data=df, palette='viridis')
plt.title(' Accidents by U.S. Region')
plt.xlabel('Region')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

print(" All Visualizations Complete! ")
```

```
"""
Robust Sentiment Analysis Script (handles EmptyDataError / empty files)
- Finds a non-empty CSV / Excel in current dir (ignores empty files)
- Tries multiple separators and encodings if needed
- Falls back to downloading from GitHub Task 4 raw folder (attempts several candidate names)
- Runs VADER sentiment analysis and produces plots + labeled CSV

Run in a Python env with:
pip install pandas matplotlib seaborn nltk wordcloud requests scikit-learn openpyxl
"""

import os
import re
import io
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
from wordcloud import WordCloud
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS

# -------------------------
# Ensure nltk vader lexicon is present
# -------------------------
try:
    nltk.data.find("sentiment/vader_lexicon.zip")
except LookupError:
    print("Downloading vader_lexicon...")
    nltk.download("vader_lexicon")

# -------------------------
# Helper: attempt to read a CSV with multiple separators/encodings
# -------------------------
def try_read_csv(path):
    # common separators and encodings
    seps = [',',';','\t','|']
    encodings = [None, 'utf-8', 'latin1', 'iso-8859-1']
    for enc in encodings:
        for sep in seps:
            try:
                # try with engine='python' to allow weird separators
                df = pd.read_csv(path, sep=sep, encoding=enc, engine='python', low_memory=False)
                # if dataframe has at least one column, accept
                if df.shape[1] >= 1 and df.shape[0] >= 1:
                    return df
            except pd.errors.EmptyDataError:
                # empty file; propagate by returning None
                return None
            except Exception:
                continue
    # final attempt: let pandas infer (no sep)
    try:
        df = pd.read_csv(path, low_memory=False)
        if df.shape[1] >= 1:
            return df
    except Exception:
        return None
    return None

def try_read_excel(path):
    try:
        df = pd.read_excel(path, engine='openpyxl')
        if df.shape[1] >= 1:
            return df
    except Exception:
        return None
    return None

# -------------------------
# Find candidate data files (CSV/XLSX) and skip empty/small ones
# -------------------------
files = os.listdir()
csv_files = [f for f in files if f.lower().endswith('.csv')]
xlsx_files = [f for f in files if f.lower().endswith(('.xlsx','.xls'))]

candidate_df = None
chosen_file = None

# prefer non-empty CSVs with likely names
def file_is_nonempty(path):
    try:
        return os.path.getsize(path) > 200  # byte threshold; adjust if needed
    except Exception:
        return False

# prefer files that look like "task4" or "social" etc
preferred_keywords = ("task", "task4", "social", "tweet", "tweets", "twitter", "comments", "reviews", "task_4")

# check CSVs
if csv_files:
    # try preferred named files first (non-empty)
    pref = [f for f in csv_files if any(k in f.lower() for k in preferred_keywords) and file_is_nonempty(f)]
    other = [f for f in csv_files if file_is_nonempty(f) and f not in pref]
    ordered = pref + other
    for f in ordered:
        print("Attempting to read CSV:", f)
        df_try = try_read_csv(f)
        if df_try is None:
            print("  -> Failed (empty or unreadable). Skipping.")
            continue
        candidate_df = df_try
        chosen_file = f
        break

# if no CSV loaded, try Excel
if candidate_df is None and xlsx_files:
    for f in xlsx_files:
        if not file_is_nonempty(f):
            continue
        print("Attempting to read Excel:", f)
        df_try = try_read_excel(f)
        if df_try is None:
            print("  -> Failed (unreadable). Skipping.")
            continue
        candidate_df = df_try
        chosen_file = f
        break

# -------------------------
# If still no dataset, attempt to download from GitHub Task 4 raw folder
# -------------------------
def try_download_github_task4():
    github_raw_base = "https://raw.githubusercontent.com/Prodigy-InfoTech/data-science-datasets/main/Task%204/"
    candidate_names = [
        "social_media_task4.csv",
        "task4.csv",
        "task_4.csv",
        "social_media.csv",
        "task4_dataset.csv",
        "dataset.csv",
        "Task4.csv",
        "Task 4.csv",
        "task-4.csv"
    ]
    for name in candidate_names:
        url = github_raw_base + name
        try:
            r = requests.get(url, timeout=10)
            if r.status_code == 200 and len(r.content) > 200:
                try:
                    dfp = pd.read_csv(io.BytesIO(r.content), low_memory=False)
                    if dfp.shape[1] >= 1:
                        print(f"Downloaded and parsed {name} from GitHub raw.")
                        return dfp
                except Exception:
                    # try separators
                    for sep in [',',';','\t','|']:
                        try:
                            dfp = pd.read_csv(io.BytesIO(r.content), sep=sep, engine='python', low_memory=False)
                            if dfp.shape[1] >= 1:
                                print(f"Downloaded and parsed {name} from GitHub raw using sep='{sep}'.")
                                return dfp
                        except Exception:
                            continue
        except Exception:
            continue
    return None

if candidate_df is None:
    print("No readable local CSV/Excel found. Attempting to download from GitHub Task 4 folder...")
    candidate_df = try_download_github_task4()
    if candidate_df is not None:
        chosen_file = "downloaded_from_github_task4"
    else:
        # final helpful error with guidance
        raise FileNotFoundError(
            "No non-empty CSV/XLSX found in the working directory and download from GitHub failed.\n"
            "Please upload the dataset CSV/XLSX to the working directory and try again."
        )

# -------------------------
# We have a dataframe in candidate_df  proceed to sentiment analysis
# -------------------------
df = candidate_df.copy()
print("Loaded dataframe shape:", df.shape)
if chosen_file:
    print("Source file:", chosen_file)

# -------------------------
# Heuristics to detect text/date/topic columns
# -------------------------
text_column_candidates = ['text', 'tweet', 'content', 'message', 'review', 'post', 'comment', 'body']
date_column_candidates = ['date', 'created_at', 'timestamp', 'time', 'datetime']
topic_column_candidates = ['topic', 'brand', 'hashtag', 'category', 'subject']

columns_lower = [c.lower() for c in df.columns]

text_col = None
for cand in text_column_candidates:
    if cand in columns_lower:
        text_col = df.columns[columns_lower.index(cand)]
        break

if text_col is None:
    # fallback: pick first object column with long median length
    obj_cols = df.select_dtypes(include=['object']).columns.tolist()
    if obj_cols:
        lengths = {}
        for c in obj_cols:
            try:
                lengths[c] = df[c].astype(str).dropna().map(len).median()
            except Exception:
                lengths[c] = 0
        text_col = max(lengths, key=lengths.get)
        print(f"Falling back to column '{text_col}' as text column (largest median string length).")
    else:
        raise ValueError("No object/text columns detected in the dataset. Please provide a CSV with a text column.")

# date and topic detection
date_col = None
for cand in date_column_candidates:
    if cand in columns_lower:
        date_col = df.columns[columns_lower.index(cand)]
        break

topic_col = None
for cand in topic_column_candidates:
    if cand in columns_lower:
        topic_col = df.columns[columns_lower.index(cand)]
        break

print("Using text column:", text_col)
if date_col:
    print("Detected date column:", date_col)
if topic_col:
    print("Detected topic/brand column:", topic_col)

# -------------------------
# Text cleaning
# -------------------------
import re
def clean_text(s):
    if pd.isna(s):
        return ""
    s = str(s).lower()
    s = re.sub(r'http\S+|www\.\S+', '', s)  # remove urls
    s = re.sub(r'@\w+', '', s)               # remove mentions
    s = re.sub(r'#', '', s)                  # drop '#', keep word
    s = re.sub(r'[^a-z0-9\s]', ' ', s)       # keep alnum + spaces
    s = re.sub(r'\s+', ' ', s).strip()
    return s

df['_clean_text'] = df[text_col].astype(str).map(clean_text)
df = df[df['_clean_text'].str.len() > 0].reset_index(drop=True)
print("After cleaning, rows with non-empty text:", df.shape[0])

# -------------------------
# Sentiment scoring with VADER
# -------------------------
sia = SentimentIntensityAnalyzer()

def vader_score(text):
    if not text:
        return {"neg":0.0,"neu":1.0,"pos":0.0,"compound":0.0}
    return sia.polarity_scores(text)

scores = df['_clean_text'].map(vader_score)
scores_df = pd.DataFrame(list(scores))
df = pd.concat([df.reset_index(drop=True), scores_df.reset_index(drop=True)], axis=1)

def label_from_compound(c):
    if c >= 0.05:
        return "positive"
    elif c <= -0.05:
        return "negative"
    else:
        return "neutral"

df['sentiment'] = df['compound'].map(label_from_compound)

# -------------------------
# Create output directories, save labeled data
# -------------------------
os.makedirs("plots", exist_ok=True)
os.makedirs("output", exist_ok=True)
out_csv = os.path.join("output", "sentiment_labeled.csv")
df.to_csv(out_csv, index=False)
print(f"Labeled dataset saved to: {out_csv}")

# -------------------------
# Visualizations
# -------------------------
sns.set(style="whitegrid")

# Sentiment distribution
plt.figure(figsize=(6,4))
order = ['positive','neutral','negative']
sns.countplot(x='sentiment', data=df, order=order)
plt.title("Sentiment Distribution (counts)")
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.tight_layout()
plt.savefig("plots/sentiment_distribution_counts.png", dpi=150)
plt.show()

plt.figure(figsize=(6,4))
(df['sentiment'].value_counts(normalize=True).reindex(order) * 100).plot(kind='bar')
plt.title("Sentiment Distribution (%)")
plt.ylabel("Percent (%)")
plt.xlabel("Sentiment")
plt.tight_layout()
plt.savefig("plots/sentiment_distribution_percent.png", dpi=150)
plt.show()

# Sentiment over time (if date present)
if date_col:
    try:
        df['_date_parsed'] = pd.to_datetime(df[date_col], errors='coerce')
        df_time = df.dropna(subset=['_date_parsed']).copy()
        df_time['_date'] = df_time['_date_parsed'].dt.date
        daily = df_time.groupby(['_date','sentiment']).size().unstack(fill_value=0).reindex(columns=order, fill_value=0)
        daily_pct = daily.div(daily.sum(axis=1), axis=0) * 100
        plt.figure(figsize=(10,5))
        daily_pct.plot.area(figsize=(12,5))
        plt.title("Sentiment Share Over Time (daily %)")
        plt.ylabel("Percent (%)")
        plt.xlabel("Date")
        plt.legend(title="Sentiment")
        plt.tight_layout()
        plt.savefig("plots/sentiment_over_time_percent.png", dpi=150)
        plt.show()
    except Exception as e:
        print("Could not parse/plot date column:", e)

# Wordcloud
from wordcloud import WordCloud
all_text = " ".join(df['_clean_text'].tolist())
stopwords = set([w.lower() for w in ENGLISH_STOP_WORDS])
def filter_tokens(text):
    return " ".join([t for t in text.split() if len(t) > 2 and not t.isnumeric() and t not in stopwords])
wc = WordCloud(width=800, height=400, background_color='white', stopwords=stopwords, collocations=True).generate(filter_tokens(all_text))
plt.figure(figsize=(10,5))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.title("Word Cloud (overall)")
plt.tight_layout()
plt.savefig("plots/wordcloud_overall.png", dpi=150)
plt.show()

# Top words by sentiment
from collections import Counter
def top_n_words(texts, n=20):
    tokens = []
    for t in texts:
        tokens += [tok for tok in t.split() if len(tok) > 2 and not tok.isnumeric() and tok not in stopwords]
    return Counter(tokens).most_common(n)

top_pos = top_n_words(df[df['sentiment']=='positive']['_clean_text'].tolist(), n=20)
top_neu = top_n_words(df[df['sentiment']=='neutral']['_clean_text'].tolist(), n=20)
top_neg = top_n_words(df[df['sentiment']=='negative']['_clean_text'].tolist(), n=20)

def plot_top_words(word_counts, title, fname):
    if not word_counts:
        return
    words, counts = zip(*word_counts)
    plt.figure(figsize=(8,4))
    sns.barplot(x=list(counts), y=list(words))
    plt.title(title)
    plt.xlabel("Count")
    plt.tight_layout()
    plt.savefig(fname, dpi=150)
    plt.show()

plot_top_words(top_pos, "Top words in Positive posts", "plots/top_words_positive.png")
plot_top_words(top_neu, "Top words in Neutral posts", "plots/top_words_neutral.png")
plot_top_words(top_neg, "Top words in Negative posts", "plots/top_words_negative.png")

# Sentiment by topic/brand (if topic_col exists)
if topic_col:
    top_topics = df[topic_col].value_counts().nlargest(8).index.tolist()
    df_topic = df[df[topic_col].isin(top_topics)]
    pivot = df_topic.pivot_table(index=topic_col, columns='sentiment', aggfunc='size', fill_value=0)[order]
    pivot_norm = pivot.div(pivot.sum(axis=1), axis=0) * 100
    pivot_norm.plot(kind='bar', stacked=True, figsize=(10,5))
    plt.ylabel("Percent (%)")
    plt.title("Sentiment share by topic/brand (top topics)")
    plt.tight_layout()
    plt.savefig("plots/sentiment_by_topic.png", dpi=150)
    plt.show()

# Top positive/negative examples
def top_examples(df_in, sentiment_label, n=5):
    subset = df_in[df_in['sentiment']==sentiment_label].copy()
    subset = subset.sort_values(by='compound', ascending=False if sentiment_label=='positive' else True)
    cols = [text_col, '_clean_text', 'compound', 'pos', 'neu', 'neg']
    cols_exist = [c for c in cols if c in subset.columns]
    return subset[cols_exist].head(n)

print("\nTop positive examples:")
print(top_examples(df, 'positive', n=5).to_string(index=False))

print("\nTop negative examples:")
print(top_examples(df, 'negative', n=5).to_string(index=False))

print("\nDone. Labeled CSV saved to 'output/sentiment_labeled.csv'. Plots saved to 'plots/'")
```
